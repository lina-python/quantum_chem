В свою очередь, по теореме Купманса для базисов 6-31G** и STO-3G они равны, соответсвенно, 0.3089 и 0.2509.

Для базиса STO-3G контрактированная функция состоит из трех гауссовых. Для водорода нужно будет 3 гауссовых функции (1s-орбиталь), для углерода и кислорода - 5 контрактированных функций на каждую орбиталь (1s, 2s, 2px, 2py, 2pz). 

C6H5OH: 6*5 + 6*1 + 5 = 41 (контрактированных)
(6*5 + 6*1 + 5)*3 = 123 (гауссовых)

C6H5NO2: 6*5 + 5*1 + 5 + 2*5 = 50 (контрактированных)
(6*5 + 5*1 + 5 + 2*5)*3 = 150 (гауссовых)

В базисе 6-31G** цифра 6 отвечает за число гауссовых функций для описания внутренних орбиталей атомов, а 3 и 1 - за описание валентного уровня двумя контрактированными функциями из трех и одной переменных. 

Смысл звездочек в добавлении к базису поляризационных функций - орбитали p-типа для атомов первого периода и орбитали d-типа для атомов второго периода. Таким образом, для водорода потребуется 12 контрактированных и 24 гауссовых функции, для атомов С 54 контрактированных и 132 гауссовых функции, для атома кислорода 9 конгтрактированных и 22 гауссовых функции. Суммарно 75 контрактированных и 178 гауссовых функций.
